### Problem 3

Mortgage payment calculations.  Write three functions. 

1. Monthly payment using the following formula.

%%latex


$$M = L\frac{I(1 + I)^N}{(1 + I)^N - 1}$$

where M = Monthly Payment, L = Loan, I = Monthly Interest, N = Number of payments


2. Write a function that calculates the total payment
3. Write a function that calculates the total interest payment
4. Write a function that calculates the interest and principal portion for all months during the lifetime of the loan.
5. Modify your code so that you can take an annualized interest rate that is monthly compounded.

```python
def monthlyPayment(L, I, N):
    a = I*((1+I)**N)
    b = ((1+I)**N)-1
    monthlyPayment = L*a/b
    return monthlyPayment
```

```python
monthlyPayment(400000,0.035/12, 12*30)
```

```python
def totalIntPayment(L, I, N):
    x = monthlyPayment(L, I, N)
    totalIntPayment = x*N
    
    return totalIntPayment
```

```python
totalIntPayment(400000,0.035/12, 12*30)
```

```python
def intPayment(L, I, N):
    x = monthlyPayment(L, I, N)
    intPayment = {}
    L_WeHave = L
    for i in range(N):
        interest = L_WeHave*I
        principalPayment = x - interest
        L_WeHave = L_WeHave-principalPayment
        print(f'Month= {i+1}, Interest = {interest}, Loan Remaining = {L_WeHave}')
        intPayment[i+1]=interest
        
    return intPayment
```

```python
def monthlyPaymentAnnual(L, I, N):
    a = (1+I/12)**(N*12)
    monthlyPayment = L*((I/12)*a/(a-1))
    return monthlyPayment
```

```python
monthlyPaymentAnnual(400000,0.035, 30)
```

```python
def totalIntPaymentAnnual(L, I, N):
    x = monthlyPaymentAnnual(L,I,N)
    totalIntPaymentAnnual = x*N*12
    return totalIntPaymentAnnual
    
```

```python
totalIntPaymentAnnual(400000,0.035, 30)
```

```python
def intPayment(L,I,N):
    x = monthlyPaymentAnnual(L,I,N)
    L_WeHave = L
    intPayment = {}
    for i in range(N*12):
        interest = L_WeHave*I/12
        principalPayment = x - interest
        L_WeHave = L_WeHave - principalPayment
        print(f'Month= {i+1}, Interest = {interest}, Loan Remaining = {L_WeHave}')
        intPayment[i+1] = interest
    return intPayment
    
```
