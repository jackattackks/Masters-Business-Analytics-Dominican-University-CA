```python
# need to join vector_df to clean df
```

```python
print(clean['complexity'].unique())
```

```python
# for complexity column

#creating instance of one-hot-encoder
complex_encoder = OneHotEncoder(handle_unknown='ignore')

#perform one-hot encoding on 'vector' column 
complex_df = pd.DataFrame(complex_encoder.fit_transform(clean[['complexity']]).toarray())

#merge one-hot encoded columns back with original DataFrame
#final_df = df.join(encoder_df)
complex_df
```

```python
complex_df.columns = ['complexity_LOW', 'complexity_HIGH']
complex_df
```

```python
# need to join complex_df to clean df
```

```python
complex_df.insert(loc=0, column='keys',value=complex_df.index)
complex_df
```

```python
print(clean['severity'].unique())
```

```python
#severity column

#creating instance of one-hot-encoder
sev_encoder = OneHotEncoder(handle_unknown='ignore')

#perform one-hot encoding on 'vector' column 
sev_df = pd.DataFrame(sev_encoder.fit_transform(clean[['severity']]).toarray())

#merge one-hot encoded columns back with original DataFrame
#final = clean.join(encoder_df)
sev_df
```

```python
sev_df.columns = ['sev_CRITICAL', 'sev_HIGH', 'sev_MEDIUM', 'sev_LOW']
sev_df
```

```python
sev_df.insert(loc=0, column='keys',value=sev_df.index)
sev_df
```

```python
# merge one hot columns to clean df using keys reference
df1 = pd.DataFrame(clean)
df2 = pd.DataFrame(vector_df)
df3 = pd.DataFrame(complex_df)
df4 = pd.DataFrame(sev_df)

merged_df1 = pd.merge(df1, df2, on='keys')
merged_df1

merged_df2 = pd.merge(df3, df4, on='keys')
merged_df2

final_merge = pd.merge(merged_df1, merged_df2, on='keys')
final_merge
```

```python
#dropping more columns that are not needed for calculation
clean_merge = final_merge.drop(columns=['keys','cve_id','vulnerability_name','vendor_project','vector','complexity','severity'])
clean_merge
```

## 5. develop classification algorithm: decision tree

```python
# data has been cleaned to include 
# product, cvss, vector, complexity, and severity up until here
```

```python
#look at how many different products can be affected
print(clean_merge['product'].unique())
```

```python
#understand the attributes of the master data frame
print("data set length:: ", len(clean_merge))
print("data set shape:: ", clean_merge.shape)
clean_merge.head()
```

```python
#import necessary packages for decision tree
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
# separate target variables from source 
# source = x
x = clean_merge.values[:, 1:11]
# target = y
y = clean_merge.values[:,0]
```

```python
# split into test and train data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)
```

```python
# function training with entropy (there's others to pick from in sklearn)
cl_entropy = DecisionTreeClassifier(criterion = 'entropy', random_state = 100,
                                   max_depth=15, min_samples_leaf=2, max_leaf_nodes=None
                                   )
cl_entropy.fit(x_train, y_train)
```

```python
# make product being affected predicition
y_pred_en = cl_entropy.predict(x_test)
y_pred_en
```

```python
#checking the accuracy
print("Model is accurate", accuracy_score(y_test, y_pred_en)*100, "percent of the time.")
```
